# To Do List App

```python
# Created an empty array to store tasks what we have to do
tasks = []

# Function to display menu for the application
def Menu():
    print("======================================")
    print("   Welcome to the To Do List App!   ")
    print("======================================")
    print("You can add, delete, and view all your tasks.\n")
    print("Let's start our program!\n")
    
    # Displaying menu options in a loop
    while True:
        print("========== Main Menu ==========")
        print("1. Add Tasks\n2. View all tasks\n3. Remove a task\n4. Exit")
        print("===============================\n")
        
        choice2 = int(input("Enter your choice (1,2,3,4) : "))        
        
        # choice for Adding tasks
        if choice2 == 1:
            addTasks()
        
        # choice for Displaying tasks
        elif choice2 == 2:
            listTasks()
        
        # choice for Removing tasks
        elif choice2 == 3:
            removeTasks()
        
        # choice for Exiting the program
        elif choice2 == 4:
            print("GoodBye!\n")
            break
            
        # if choice does not matches the required options, show invalid choice
        else:
            print("Invalid choice")
```

```python
# Function to add tasks to the list
def addTasks():
    task = input("Enter your task you want to add: ")
    tasks.append(task)
    print(f"Your Task '{task}' has been sucessfully added.\n")
```

```python
# Function to list all tasks
def listTasks():
    
    if not tasks:
        print("No tasks available.\n")
        
    else:
        print("\n=========== Your Tasks ===========")
        for i, task in enumerate(tasks, start=1):
            print(f"{i}. {task}")
        print("==================================\n")
```

```python
# Function to remove a task from the list
def removeTasks():
    if not tasks:
        print("No tasks present.\n")
        return
    
    # List tasks before asking for which one to remove
    listTasks()
    
    try:
        choice = int(input("Enter the task number you want to delete: "))
        if choice < 1 or choice > len(tasks):
            print("Please enter a valid task number.\n")
        else:
            removed_task = tasks.pop(choice - 1)
            print(f"'{removed_task}' has been successfully removed.\n")
    except ValueError:
        print("Please enter a valid number.\n")
```

```python
# Entry point of the program
if __name__ == "__main__":
    Menu()
```
